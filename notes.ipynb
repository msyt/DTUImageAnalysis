Reading the image

```python
# import skimage
from skimage import io

# read image
im_color = io.imread("metacarpals.png")

# read image as gray scale
im_gray = io.imread("metacarpals.png", as_gray=True) # This is Python. Use "True"
```

Checking the image


```python
# Dimension
print(im_color.shape)

# Pixel type
print(im_color.dtype)
```

Showing the image

```python
import matplotlib.pyplot as plt
# plt.imshow(im_color)
plt.imshow(im_gray, cmap='gray')  # Specify the 'gray' colormap explicitly
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("Hi")
plt.imshow(im_gray)
plt.imshow(im_gray, vmin=np.min(im_gray), vmax=np.max(im_gray), cmap="gray")

io.show()
```
